- DIV (배당수익률) : (주가배당금 / 주가) * 100
- BPS (주당순자산가치=청산가치) : 순자산 / 총발행주식수
- PER (주가수익비율) : 주가 / 주당순이익 
- EPS (주당순이익) : 당기순이익 / 총발행주식수 
- PBR (주가순자산비율) = 주가 / BPS = PER*EPS / BPS
