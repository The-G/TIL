```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 랜덤한 숫자 데이터 생성 (1~99999)
data = np.random.randint(1, 100000, 10000)
```

```python
data
```

```python
# 첫 번째 자리 숫자 추출
first_digits = [int(str(num)[0]) for num in data]
```

```python
first_digits
```

```python
# 첫 자리 숫자의 빈도수 계산
unique, counts = np.unique(first_digits, return_counts=True)
frequencies = counts / sum(counts) * 100
```

```python
frequencies
```

```python
# 벤포드의 기대 확률
benford_probs = [np.log10(1 + 1/d) * 100 for d in range(1, 10)]
```

```python
# 한글 폰트 설정 (예: 맑은 고딕)
plt.rc('font', family='Malgun Gothic')  # Windows 사용자의 경우
# plt.rc('font', family='AppleGothic')  # Mac 사용자의 경우

# 마이너스 (-) 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python

# 5. 그래프 시각화
plt.bar(unique, frequencies, label="관측된 데이터", alpha=0.7)  # 실제 데이터 막대 그래프
plt.plot(range(1, 10), benford_probs, marker="o", linestyle="dashed", 
         color="red", label="벤포드 법칙")  # 벤포드 법칙의 예상 확률 (점선 그래프)

# 6. 그래프 설정
plt.xlabel("첫 번째 자리 숫자")  # X축 레이블
plt.ylabel("비율 (%)")  # Y축 레이블
plt.xticks(range(1, 10))  # X축 눈금 설정 (1~9)
plt.legend()  # 범례 표시
plt.title("벤포드 법칙 분석")  # 그래프 제목
plt.show()  # 그래프 출력
```

- 결과 해석 
    - 벤포드 법칙과 차이 
        - 벤포드 법칙에 따르면 1로 시작하는 숫자가 가장 많이 나와야 하지만, 
        - 현재 관측된 데이터에서는 모든 숫자가 비슷한 빈도를 가지며, 균등한 분포를 보임 
        - 즉, 벤포드 법칙을 따르지 않는 데이터입니다. 
    - 이유 분석 
        - 벤포드 법칙은 "자연 발생 데이터"에서 성립합니다. 
        - 하지만 현재 데이터는 균등 분포에서 무작위로 생성한 숫자이므로 벤포드 법칙을 따르지 않습니다.
